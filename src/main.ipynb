```python
from data.data_ingestion import DataIngestion
from preprocessing.data_preprocessing import Preprocessor
from features.feature_engineering import FeatureEngineer
from inference.prediction import Predictor
```

```python
ingestor = DataIngestion(spark)
df_raw = ingestor.load_table("samples.nyctaxi.trips", limit=1)
```

```python
preprocessor = Preprocessor()
df_clean = preprocessor.clean(df_raw)
```

```python
featureengineer = FeatureEngineer()
psdf = featureengineer.add_time_features(df_clean)
psdf = psdf.to_pandas()
```

```python
predictor = Predictor()
pred = predictor.predict(psdf)
pred = round(pred[0], 2)
```

```python
print(f"Prediction: {pred}")
```
